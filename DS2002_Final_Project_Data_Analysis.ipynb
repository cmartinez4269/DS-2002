```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('/content/sample_data/transformed_data.csv')

df.rename(columns={'year_of_ozone_and_co2': "Year"}, inplace=True)
df.rename(columns={'mean_of_ozone_and_co2': "Mean of Ozone and CO2"}, inplace=True)
print(df.head())

fig, ax1 = plt.subplots()
ax1.plot(df['Year'], df['mean_of_ozone'], label='Ozone', color='blue')
ax1.set_xlabel('Year')
ax1.set_xlabel('Mean (PPM)')
plt.title('Mean Ozone and CO2 Levels Over Marine Surface Sites')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
ax2.plot(df['Year'], df['mean'], label='CO2', color='black')
ax2.set_ylabel('Mean (PPM)')
ax2.tick_params(axis='y', labelcolor='black')

plt.title('Mean Ozone and CO2 Levels Over Marine Surface Sites')
plt.show()
```

```python
#calculating correlation between ozone and co2 means
correlation = df['mean_of_ozone'].corr(df['mean'])
print(correlation)
```

```python
#calculate average mean of ozone values
mean_mean_ozone = df['mean_of_ozone'].mean()
print(f"The mean of ozone is: {mean_mean_ozone}")
```
